```python
#Ejercicio 1
sns.histplot(x = "birth_rate", data = df, color = "violet");

plt.xlabel(" Nacimientos ")
plt.ylabel("Recuento")
plt.title("Tasa de natalidad", fontsize = 10);

# La tasa de natalidad se refiere al numero de nacimientos vivos durante un año, la mayor tasa esta entre 5/10 niños por años
```

```python
#Ejercicio 2
sns.regplot(x = "birth_rate", 
                y = "life_expectancy", 
                data = df,
                line_kws = {"color": "black", "linewidth": 1},
                color = "darkcyan")


plt.xlabel("Nacimientos")
plt.ylabel("Esperanza de vida ")
plt.title("Relacion nacimientos/esperanza", fontsize = 10);

# A mayor esperanza de vida menos numero de nacimientos.
```

```python
# 3. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

df_paises = df.groupby("continent")["country"].count().reset_index()
df_paises
```

```python
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"]
plt.pie("country", labels= "continent",
        data = df_paises, 
        autopct=  '%1.1f%%', 
        colors = colores, 
        textprops={'fontsize': 8}, 
        startangle=90);

plt.xlabel("Continentes")
plt.title("Relacion paises", fontsize = 10);

# El continente con mayor numero de paises es Africa
```

```python
# 4. ¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos. 

df_paises2 = df.loc[df["continent"] == "Europe"]
df_paises2.head()
```

```python
sns.boxplot(y = "life_expectancy", 
            data = df_paises2, 
            width = 0.5, 
            color = "turquoise")


plt.ylabel("Esperanza de vida ")
plt.title("Esperanza de vida en Europa", fontsize = 10);

# La esperanza de vida esta entorno a los 80 años
```

```python
#5. ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5))
#axes=axes.flat # Aplana los ejes y podemos acceder y no tenemos que entrar por posicion


sns.regplot(x = "birth_rate", y = "infant_mortality", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1},
            scatter_kws = {"color": "teal", "s": 1}, 
            ax = axes[0][0]
            )

sns.regplot(x = "birth_rate", y = "life_expectancy", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1},
            scatter_kws = {"color": "teal", "s": 1}, 
            ax = axes[0][1]
            )

sns.regplot(x = "birth_rate", y = "density", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1},
            scatter_kws = {"color": "teal", "s": 1}, 
            ax = axes[1][0]
            )

sns.regplot(x = "birth_rate", y = "urban_population", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1},
            scatter_kws = {"color": "teal", "s": 1}, 
            ax = axes[1][1]
            )


# las graficas de motartalidad y natalidad son proporcionales 
# Esperanza de vida y natalidad son inversamente proporcionales (a mayor esperanza de vida menor tasa de natalidad)
# La densidad se mantiene estable con la natalidad igual que con la grafica de poblacion urbana
```

```python
# Ejercicio 6 ¿Cuáles son los 10 países con la mayor área forestal?


df_paises_3 = df.groupby("country")["forested_area"].sum().reset_index().sort_values(by="forested_area", ascending=False)
top_ten_forest= df_paises_3[:10]

sns.countplot(x = "forested_area", 
              data = top_ten_forest, 
              palette = "viridis", 
              hue = "country")


```

```python
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen", "pink", "orange", "black"]

plt.pie("forested_area", labels= "country",
        data = top_ten_forest, 
        autopct=  '%1.1f%%', 
        colors = colores, 
        textprops={'fontsize': 8}, 
        startangle=90);
```
